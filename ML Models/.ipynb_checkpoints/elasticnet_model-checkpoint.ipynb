```python
import pandas as pd
import numpy as np
from numpy.random import seed
seed(42)
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

## Data Preprocessing

```python
# Read in movie csv
movies = pd.read_csv("../Resources/imdb_final.csv")
movies.head()
```

```python
# Set vote column to integer type
movies["median_vote"] = movies["median_vote"].astype(int)
```

```python
# Set values for X and y
X = movies[["year", "genre", "duration", "budget"]]
y = movies["median_vote"].values.reshape(-1,1)
print(X.shape, y.shape)
```

```python
data = X.copy()
data
```

```python
# Dummy Encoding for genre column
data_binary_encoded = pd.get_dummies(data, columns= ["genre"])
data_binary_encoded.head()
```

```python
# Create train/test split
X_train, X_test, y_train, y_test = train_test_split(data_binary_encoded, y, random_state=42)
```

```python
# Create StandardScaler model and fit to training data
X_scaler = StandardScaler().fit(X_train)
# y_scaler = StandardScaler().fit(y_train)

# Transform training and testing data using X_scaler and y_scaler
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
# y_train_scaled = y_scaler.transform(y_train)
# y_test_scaled = y_scaler.transform(y_test)
# print(y_train_scaled)
```

## Create Elastic Net Regression Model

```python
# Create elasiticnet regression model
en_model = ElasticNet(alpha= 0.01).fit(X_train_scaled, y_train)
```

```python
training_score = en_model.score(X_train_scaled, y_train)
testing_score = en_model.score(X_test_scaled, y_test)
print(f"Linear Regression Training Score: {training_score: .3f}")
print(f"Linear Regression Testing Score: {testing_score: .3f}")
```

## Grid Search

```python
from sklearn.model_selection import GridSearchCV
```

## Plot Residuals

```python
predictions_train = en_model.predict(X_train)
predictions_test = en_model.predict(X_test)

plt.scatter(predictions_train, predictions_train - y_train, c="blue", label= "Training Data")
plt.scatter(predictions_test, predictions_test - y_test, c="red", label= "Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y.min(), xmax=y.max())
plt.title("Residual Plot")
plt.show()
```
